```python
import pandas as pd
```

```python
seoul_df = pd.read_csv('datas/seoul.csv', encoding='cp949')
```

```python
seoul_df.drop(['2024년_계_연령구간인구수', '2024년_계_0~9세','2024년_계_10~19세','2024년_계_20~29세','2024년_계_30~39세','2024년_계_40~49세','2024년_계_50~59세'], axis=1, inplace=True)
```

```python
seoul_df = seoul_df.loc[[0, 21]]
```

```python
seoul_df.dtypes
```

```python
# 합산할 대상 컬럼들
age_columns = [
    '2024년_계_60~69세',
    '2024년_계_70~79세',
    '2024년_계_80~89세',
    '2024년_계_90~99세',
    '2024년_계_100세 이상'
]

# 문자열 -> 쉼표 제거 -> 정수로 변환
for col in age_columns:
    seoul_df[col] = seoul_df[col].astype(str).str.replace(',', '').astype(int)

# 새로운 열 '노령인구_합계' 생성
seoul_df['old_sum'] = seoul_df[age_columns].sum(axis=1)
```

```python
seoul_df['2024년_계_총인구수'] = seoul_df['2024년_계_총인구수'].astype(str).str.replace(',', '').astype(int)
```

```python
seoul_df['행정구역'] = seoul_df['행정구역'].replace({
    '서울특별시 (1100000000)': '서울시',
    '서울특별시 관악구 (1162000000)': '관악구'
})
```

```python
seoul_df['행정구역'] = seoul_df['행정구역'].str.replace(r'\n?\(.*\)', '', regex=True)
```

```python
seoul_df['행정구역'] = seoul_df['행정구역'].replace({'서울특별시': '서울시'})
```

```python
import matplotlib.pylab as plt
# '-' 깨짐 해결 모듈
import matplotlib as mpl

plt.rcParams["font.family"] = "Malgun Gothic"  # For Windows
plt.rcParams["font.size"] = 12
plt.rcParams["figure.figsize"] = (8,4)
print(plt.rcParams["font.family"])

# 마이너스 깨짐 해결
mpl.rcParams["axes.unicode_minus"] = False
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 데이터 구성
data = {
    '행정구역': ['서울특별시', '관악구'],
    'old_sum': [2506338, 119602]
}
df = pd.DataFrame(data)

# 비율 계산
df['노령인구비율(%)'] = df['old_sum'] / df['old_sum'].iloc[0] * 100

# 시각화 (비율 기준)
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='행정구역', y='노령인구비율(%)', palette='pastel')

# 텍스트 표시
for i in range(len(df)):
    plt.text(i, df['노령인구비율(%)'][i] + 1,
             f"{df['노령인구비율(%)'][i]:.2f}%",
             ha='center', fontsize=12, fontweight='bold')

plt.title('서울특별시 대비 관악구 노령인구 비율 비교')
plt.ylabel('서울시 대비 노령인구 비율 (%)')
plt.xlabel('')
plt.ylim(0, max(df['노령인구비율(%)']) + 10)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 데이터 구성
data = {
    '행정구역': ['서울시', '관악구'],
    '노령인구비율(%)': [26.85, 25.03]
}
df = pd.DataFrame(data)

# 색상 지정 (서울시는 회색, 관악구는 하늘색)
colors = ['gray', 'skyblue']

# 시각화
plt.figure(figsize=(8, 6))
sns.barplot(data=df, x='행정구역', y='노령인구비율(%)', palette=colors)

# 막대 위 텍스트 표시
for i, value in enumerate(df['노령인구비율(%)']):
    plt.text(i, value + 0.5,
             f"{value:.2f}%", 
             ha='center',
             fontsize=12,
             color='blue' if i == 1 else 'black',
             fontweight='bold')

# 제목 및 축 라벨 설정
plt.title('서울시 대비 관악구 노령인구 비율 비교')
plt.ylabel('노령인구 비율 (%)')
plt.ylim(0, max(df['노령인구비율(%)']) + 5)
plt.xlabel('')
plt.tight_layout()
plt.show()
```
